```python
# second try for spans

#create dictionaries for the markables and the words
markables = {}
for m in tree_dianette:
    if "id" in m.attrib:
        markables[m.attrib["id"]] = m.attrib

words = {}
for w in branch_dianette.iter():
    tag = w.tag
    if tag.endswith("word") and "id" in w.attrib:
        words[w.attrib["id"]] = w.text


#since we use the spans, in the data the spans contain the word ids written in different forms of puctuation based on their lengths
#therefroe we create a function to handle theese lengths and caount the real spans in order to map them later 
def span_to_text(span_str, words_dict):
    if not span_str:
        return "?"
    
    word_ids = []
    parts = span_str.split(',')
    
    for part in parts:
        part = part.strip()
        if '..' in part:
            try:
                start_id, end_id = part.split('..')
                start = int(start_id.replace("word_", ""))
                end = int(end_id.replace("word_", ""))
                word_ids.extend(f"word_{i}" for i in range(start, end + 1))
            except ValueError:
                word_ids.append("?")
        else:
            word_ids.append(part) 
    
    return " ".join(words_dict.get(wid, "?") for wid in word_ids)

# colelct clustered markalbes  except the abstract ones
coref_clusters = {}
for mid, info in markables.items():
    if info.get("category") == "abstract":
        continue
    coref_set = info.get("coref_set")
    ref_type = info.get("ref_type")
    if coref_set and ref_type == "phrase":
        coref_clusters.setdefault(coref_set, []).append(mid)

#trace the antecedents for each markable id and pays also attention to split antecedents
def trace_antecedents(mid):
    chain = []
    current = mid
    seen = set()
    while current and current not in seen:
        seen.add(current)
        chain.append(current)
        info = markables.get(current)
        next_ = info.get("single_phrase_antecedent") or (
            info.get("multiple_phrase_antecedents", "").split(";")[0]
            if info.get("multiple_phrase_antecedents") else None
        )
        current = next_
    return list(reversed(chain))

#vizualize wha we have yet 
#firstly sort them out 
for set_id, mids in coref_clusters.items():
    collected = set(mids)
    for mid in mids:
        collected.update(trace_antecedents(mid))
    mids_sorted = sorted(collected, key=lambda m: int(m.replace("markable_", "")))

# turn them to real words from word_ids, but we also keep the word ids for validation 
    readable = []
    for mid in mids_sorted:
        info = markables[mid]
        span = info.get("span")
        real_text = span_to_text(span, words) if span else "?"
        readable.append(f"{mid} (span: {span}) → {real_text}")

    if readable:
        print(f"\nSet {set_id}: " + " > ".join(readable))
```

```python
model_version = 'bert-base-uncased' # similarly to before the mdels are changed here acoordingly  
model = BertModel.from_pretrained(model_version, output_attentions=True)
tokenizer = BertTokenizerFast.from_pretrained(model_version) #the same as before 
```

```python
# since we dont have the raw text aligned to the data we concatonate the words for the raw text
sorted_ids = sorted(words, key=lambda w: int(w.replace("word_", "")))
raw_text = " ".join(words[wid] for wid in sorted_ids)
print(raw_text) 
```

```python
#model parameters

model_name = "bert-base-uncased"
fast_tokenizer = BertTokenizerFast.from_pretrained(model_name)
model = model.to("cuda" if torch.cuda.is_available() else "cpu") #push to gpu 
model.eval()

# tokenization
encodings = fast_tokenizer(
    raw_text,
    return_offsets_mapping=True,
    return_tensors='pt',
    truncation=True,
    padding='max_length',
    max_length=512,
    stride=256,
    return_overflowing_tokens=True,
    return_special_tokens_mask=True
)

# create some lists for the indices, the mappinig positions and the embeddings 
all_embeddings = []
all_offset_mappings = []
all_input_ids = []


#turnt thme to embeddings
with torch.no_grad():
    for i in range(len(encodings['input_ids'])):
        input_ids = encodings['input_ids'][i].unsqueeze(0).to(model.device)
        attention_mask = encodings['attention_mask'][i].unsqueeze(0).to(model.device)
        offsets = encodings['offset_mapping'][i].tolist()

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        emb = outputs.last_hidden_state.squeeze(0)

        all_embeddings.append(emb.cpu())
        all_offset_mappings.append(offsets)
        all_input_ids.append(input_ids.squeeze(0).cpu())
```

```python
# since we have utilized the offset mapping we need to alignt he character levels to words and then to postions in the text

#we start the charatcter counting for builing words 
word_char_spans = {} 
char_cursor = 0
word_ids_sorted = sorted(words.keys(), key=lambda x: int(x.split("_")[1]))

for wid in word_ids_sorted:
    token = words[wid]
    if token is None:
        continue
    token = token.strip() # create tokens from whitespace
    if not token:
        continue

    # Try to locate the token from the current position
    idx = raw_text.find(token, char_cursor)

    # if idx == -1:
    #     print(f"Warning: Token '{token}' (word ID {wid}) not found in text.")
    #     continue

    start_char = idx
    end_char = idx + len(token)
    word_char_spans[wid] = (start_char, end_char)
    #continue to the next one 
    char_cursor = end_char
```

```python
# create a dictionary to map the charachters to tokens 
char_to_token_index = {}  

#map and position the token chucks  with  the offsets in the sentences 
for chunk_id, offsets in enumerate(all_offset_mappings):
    for token_id, (start, end) in enumerate(offsets):
        if start == 0 and end == 0:
            continue  
        char_to_token_index[(start, end)] = (chunk_id, token_id)
```

```python
# align the word embeddings

word_to_embedding = {}

#we use the token positioning and then retrieve the tokens this part was corrected with the help of GPT 
for word_id, (start_char, end_char) in word_char_spans.items():
    token_pos = char_to_token_index.get((start_char, end_char))
    if token_pos:
        chunk_id, token_id = token_pos
        word_to_embedding[word_id] = all_embeddings[chunk_id][token_id] #embs from the hidden states

# print(f"Total words IDs with embeddings: {len(word_to_embedding)}")
```

```python
# previously the problem with BER is that it takes as input sequence the maximum of only 512 tokens. Therefore we must have do the chucking method like (Joshi et al., 2019)
#in order to build the similartity chains across all documents. 

from collections import defaultdict

#create a dictionary and iterrate throught the word's offsets for the positions of the tokens
word_char_spans = {}
current_offset = 0
for i, wid in enumerate(sorted(words.keys(), key=lambda x: int(x.replace("word_", "")))):
    word = words[wid]
    if word is None:
        continue
    idx = raw_text.find(word, current_offset)
    if idx == -1:
        continue
    word_char_spans[wid] = (idx, idx + len(word))
    current_offset = idx + len(word)

# map the tokens from offset
word_to_token_indices = defaultdict(list)

# itterate through token chucks for capturing the full text #help of GPT
for chunk_id, offsets in enumerate(all_offset_mappings):
    for token_idx, (start, end) in enumerate(offsets):
        token_abs_idx = chunk_id * 512 + token_idx
        for wid, (w_start, w_end) in word_char_spans.items():
            # Include token if it's fully inside the word span
            if w_start <= start and end <= w_end:
                word_to_token_indices[wid].append(token_abs_idx)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# we concatenatt the emeddings to one tensor 
full_embeddings = torch.cat(all_embeddings, dim=0)

# we turnt the spans of tokens to embeddings, similarly as before we have to take into consideration the different type of spans
# with the difference like ",", "..",(indicating that there is 2 or more words) or one simple head noun that might be used as a span
def get_embedding_for_span(span):
    if not span:
        return None

    word_ids = []

    for part in span.split(','):
        part = part.strip()
        if '..' in part:
            try:
                start_str, end_str = part.split('..')
                start = int(start_str.replace("word_", ""))
                end = int(end_str.replace("word_", ""))
                word_ids.extend([f"word_{i}" for i in range(start, end + 1)])
            except ValueError:
                continue
        else:
            word_ids.append(part)

    # ma the tokens that are nouns from other random tokens
    token_indices = []
    for wid in word_ids:
        if wid in word_to_token_indices:
            token_indices.extend(word_to_token_indices[wid])

    if not token_indices:
        return None

    # average the span embeddings 
    return full_embeddings[token_indices].mean(dim=0)


# we go thrpough the chains that we have created from the previous step
for set_id, mids in coref_clusters.items():
    collected = set(mids)
    for mid in mids:
        collected.update(trace_antecedents(mid))
    mids_sorted = sorted(collected, key=lambda m: int(m.replace("markable_", "")))

    print(f"\nChain {set_id}:")
    prev_embedding = None
    for mid in mids_sorted: #visualize it 
        span = markables[mid].get("span")
        text = span_to_text(span, words)
        emb = get_embedding_for_span(span)

        #help of GPT
        #mesure the embeddings similarities for each step
        if emb is None:
            print(f"{mid}: \"{text}\" [no embedding]")
            continue

        if prev_embedding is None:
            print(f"{mid}: \"{text}\" [First in chain]")
        else:
            score = cosine_similarity(
                emb.unsqueeze(0),
                prev_embedding.unsqueeze(0) #similarities between spans
            )[0][0].item()
            print(f"{mid}: \"{text}\" → Cosine similarity to previous: {score:.4f}")

        prev_embedding = emb
```

```python
# we map oncce again the ids t the words
def make_word_lookup(df_words):
    return dict(zip(df_words["id"], df_words["word"]))

# align the words in the spans # help of chat
def expand_span(span): 
    if ".." in span:
        s, e = span.split("..")
        s_i = int(s.split("_")[1]); e_i = int(e.split("_")[1])
        return [f"word_{i}" for i in range(s_i, e_i + 1)]
    return [span]


def extract_words_from_span_df(dfsent, word_lookup): #sentece spans
    out = dfsent.copy() #copy to be safe
    out["word_ids"] = out["span"].apply(expand_span) # spans to word ids
    out["sentence"] = out["word_ids"].apply( #word_ids to words
        lambda wids: " ".join(word_lookup.get(w, "") for w in wids)
    )
    #into dataframe
    return out[["id","sentence","span","word_ids"]].rename(columns={"id":"sent_id"})
```

```python
#the creation fo the positive and negative examples was done eith the assistance of Ludmila Bajuk

def build_coref_flat(dfcoref: pd.DataFrame, doc_id: str | None = None) -> pd.DataFrame: #create the dataframe

    ids, spans, referrings = [], [], []

    # get the columns that we want 
    for _, ref in dfcoref.iterrows():
        ref_type = ref.get("reference")
        mid = ref.get("id")
        span = ref.get("span")
        ant = ref.get("single_phrase_antecedent")

        # prefix current markable id
        if doc_id and isinstance(mid, str):
            mid = f"{doc_id}_{mid}"


        # distinguish the new antecedetns to the ones that have been previously introduced as or if they areferring back to somewhere else
        if ref_type == "new":
            ids.append(mid); spans.append(span); referrings.append("new")
        elif ref_type == "old":
            # normalize antecedents # help of chat 
            if isinstance(ant, str) and ant.strip():
                if doc_id:
                    ant = ";".join(f"{doc_id}_{p.strip()}" for p in ant.split(";") if p.strip())
                else:
                    ant = ";".join(p.strip() for p in ant.split(";") if p.strip())
                ids.append(mid); spans.append(span); referrings.append(ant)
            else:
                # keep row but mark as None if you want to filter later
                ids.append(mid); spans.append(span); referrings.append(None)

    out = pd.DataFrame({"id": ids, "span": spans, "referring": referrings})
    return out
```

```python
def build_coref_flat(dfcoref): #words to the data
    """
    extract id, span and referring markables from a coref DataFrame and 
    keep both 'new' and 'old' references, but for 'old' uses single_phrase_antecedent.
    """
    ids = []
    spans = []
    referrings = []

    # go through the markables 
    for _, ref in dfcoref.iterrows():
        if ref["reference"] == "new":
            ids.append(ref["id"])
            spans.append(ref["span"])
            referrings.append("new")
        elif ref["reference"] == "old": #old entotoes are trace to antecedents
            ids.append(ref["id"])
            spans.append(ref["span"])
            if ref["single_phrase_antecedent"]:
                referrings.append(ref["single_phrase_antecedent"])
            else:
                referrings.append(None)

    df_new = pd.DataFrame({
        "id": ids,
        "span": spans,
        "referring": referrings
    })

    return df_new
```

```python
def expand_span_safe(span: str): #handle cases that arent only dots for spans better than befor 
    if not isinstance(span, str) or not span.strip():
        return []
    ids = []
    for chunk in re.split(r'[;,]\s*', span.strip()):
        if not chunk:
            continue
        nums = re.findall(r'word_(\d+)', chunk)
        if '..' in chunk:
            if len(nums) >= 2:
                a, b = int(nums[0]), int(nums[1])
                lo, hi = (a, b) if a <= b else (b, a)
                ids.extend([f"word_{i}" for i in range(lo, hi+1)])
            elif len(nums) == 1:
                ids.append(f"word_{nums[0]}")
        else:
            ids.extend([f"word_{int(n)}" for n in nums])
    # de-dup preserve order
    seen, out = set(), []
    for wid in ids:
        if wid not in seen:
            seen.add(wid); out.append(wid)
    return out

def build_coref_with_words(dfcoref, df_words): # add the words like before 
    word_lookup = dict(zip(df_words["id"], df_words["word"]))


    #sspant to text function again
    def span_text(span):
        wids = expand_span_safe(span)
        return " ".join(word_lookup.get(wid, "") for wid in wids)


    # add the extra columns that we need for the tasks
    ids, spans, referrings, words, gram_fncs = [], [], [], [], []
    for _, ref in dfcoref.iterrows():
        rtype = ref.get("reference")
        mid = ref.get("id")
        span = ref.get("span")
        gram = ref.get("gram_fnc")
        if gram not in {"subj", "obj"}:
            gram = "other"
        if rtype == "new":
            ids.append(mid); spans.append(span); referrings.append("new")
            words.append(span_text(span)); gram_fncs.append(gram)
        elif rtype == "old":
            ant = ref.get("single_phrase_antecedent")
            ids.append(mid); spans.append(span); referrings.append(ant if ant else None)
            words.append(span_text(span)); gram_fncs.append(gram)
    
    #create the new dataframe
    df_new = pd.DataFrame({
        "id": ids,
        "span": spans,
        "referring": referrings,
        "words": words,
        "gram_fnc": gram_fncs
    })

    df_new = df_new[df_new["referring"].notna()].reset_index(drop=True)
    return df_new
```

```python
#postive examples

def build_positives_from_coref(df_new: pd.DataFrame, df_s: pd.DataFrame) -> pd.DataFrame:
    """
    positive pairs, with pronominal references, excluding pronouns as atecedents 

    """
    regex = r"(markable_\d+)(;(markable_\d+))?"

    #get the first word of the sentence span 
    def first_word(span):
        m = re.match(r"(word_\d+).*", span or "")
        return m.group(1) if m else None

    #get the word withing a sequence
    def word_index(word_id):
        if not word_id:
            return None
        return int(word_id.replace("word_", ""))

    #map the sentences based on words
    def sentence_for_word(word_id):
        if not word_id:
            return None
        for _, row in df_s.iterrows():
            if isinstance(row["word_ids"], list) and word_id in row["word_ids"]:
                return row["sentence"]
        return None

    #create new df lists
    sent1, sent2, antecedent, reference = [], [], [], []
    ant_ids, ref_ids, cos_sims, ant_fncs = [], [], [], []
    antecedent_embeddings, reference_embeddings = [], []

    #we check here if there is one antecedent and pass it or if there is an other one also 
    refs_list = df_new["referring"].tolist()
    for _, row1 in df_new.iterrows():
        for i, ref in enumerate(refs_list):
            match1 = re.match(regex, ref or "")
            if not match1:
                continue


            #deal with the split antecedents
            if row1["id"] == match1.group(1) or (
                match1.group(2) and ";" + row1["id"] == match1.group(2)
            ):
                second = df_new.iloc[i]

                #pronouns as references
                if second["words"].lower().strip() not in pronouns:
                    continue

                # avoid pronouns as antecedents
                if row1["words"].lower().strip() in pronouns:
                    continue

                # antecedent before reference
                ant_word = word_index(first_word(row1["span"]))
                ref_word = word_index(first_word(second["span"]))
                if ant_word is None or ref_word is None:
                    continue
                if ant_word >= ref_word:
                    continue

                # build the pairs
                ante_text = row1["words"]
                ref_text  = second["words"]
                ante_span = row1["span"]
                ref_span  = second["span"]


                #get the embeddings
                ante_emb = get_embedding_for_span(ante_span)
                ref_emb  = get_embedding_for_span(ref_span)
                if ante_emb is None or ref_emb is None:
                    continue

                #get cosine similarity
                cos_sim = cosine_similarity(
                    ante_emb.unsqueeze(0),
                    ref_emb.unsqueeze(0)
                )[0][0].item()

                #append the things to the list
                antecedent.append(ante_text)
                reference.append(ref_text)
                ant_ids.append(row1["id"])
                ref_ids.append(second["id"])
                sent1.append(sentence_for_word(first_word(row1["span"])) or "")
                sent2.append(sentence_for_word(first_word(second["span"])) or "")
                cos_sims.append(cos_sim)
                antecedent_embeddings.append(ante_emb.detach().cpu().numpy())
                reference_embeddings.append(ref_emb.detach().cpu().numpy())
                ant_fncs.append(row1.get("gram_fnc", None))

    return pd.DataFrame({
        "Sent_1": sent1,
        "Sent_2": sent2,
        "antecedent": antecedent,
        "reference": reference,
        "antecedent_id": ant_ids,
        "reference_id": ref_ids,
        "antecedent_function": ant_fncs,
        "cosine_sim": cos_sims,
        "embedding_ante": antecedent_embeddings,
        "embedding_ref": reference_embeddings,
        "label": 1
    })
```

```python
def is_valid_antecedent(text: str) -> bool:
    """
    make sure that the antecedents are valid 
    """
    if not text or not isinstance(text, str):
        return False
    t = text.lower().strip()
    # exclude second-person and other unwanted forms
    bad = {"you", "your", "yours", "yourself", "yourselves"}
    if t in bad:
        return False
    return True
```

```python
#negative examples 

# the previous span extraction didnt work properly so we included some with helpers witht he help of GPT
def expand_span_safe(span: str):
    if not isinstance(span, str) or not span.strip():
        return []
    ids = [] #new id lsit
    for chunk in re.split(r'[;,]\s*', span.strip()): # split spans
        if not chunk:
            continue
        nums = re.findall(r'word_(\d+)', chunk) #extract them and define more delicate if there are 1, 2 or more to count them and add them in a list 
        if '..' in chunk:
            if len(nums) >= 2:
                a, b = int(nums[0]), int(nums[1])
                lo, hi = (a, b) if a <= b else (b, a)
                ids.extend([f"word_{i}" for i in range(lo, hi+1)])
            elif len(nums) == 1:
                ids.append(f"word_{nums[0]}")
        else:
            ids.extend([f"word_{int(n)}" for n in nums])
    seen, out = set(), []
    for wid in ids:
        if wid not in seen:
            seen.add(wid); out.append(wid)
    return out

#find the sentences, the words and the correct sets 
def build_sentence_index(df_s: pd.DataFrame):
    wid2sid, sid2text, sid2set = {}, {}, {}
    for _, r in df_s.iterrows():
        sid2text[r["sent_id"]] = r["sentence"] #sentence id to text
        sset = set(r["word_ids"])
        sid2set[r["sent_id"]] = sset # sentence to words 
        for wid in r["word_ids"]:
            wid2sid[wid] = r["sent_id"] # single words to sentences
    return wid2sid, sid2text, sid2set


#negative examples 
def build_markables_for_negatives(df_new: pd.DataFrame, df_s: pd.DataFrame) -> pd.DataFrame: 
    #apply the new function to our previous files with the words sentences and info
    wid2sid, sid2text, sid2set = build_sentence_index(df_s)

    df_m = df_new.copy() #copy for safety
    df_m["word_ids"] = df_m["span"].apply(expand_span_safe) 
    df_m["word_ids_set"] = df_m["word_ids"].apply(set)

    #find the sentence from the word
    def first_word(span):
        m = re.search(r"(word_\d+)", span or "")
        return m.group(1) if m else None

    sent_ids, sent_texts, sent_sets = [], [], []
    for _, r in df_m.iterrows():
        w0 = first_word(r["span"])
        sid = wid2sid.get(w0)
        sent_ids.append(sid)
        sent_texts.append(sid2text.get(sid))
        sent_sets.append(sid2set.get(sid, set()))

    df_m["sent_id"] = sent_ids
    df_m["sent_text"] = sent_texts
    df_m["sent_words_set"] = sent_sets
    return df_m
def build_negatives_from_coref(df_pos: pd.DataFrame,
                               df_m: pd.DataFrame,
                               pronouns: set,
                               markables: dict,
                               full_coref_clusters=None) -> pd.DataFrame:
    """
    based on the  golden reference anfd the golden NP we search for other NPs meaning makrables close
    to the golden antecedent. the shouldnt be pronouns or belong to the same cluster of references
    """
    # markables to clusters 
    markable2cluster = {}
    if full_coref_clusters:
        for cid, mids in full_coref_clusters.items():
            for m in mids:
                markable2cluster[m.strip()] = cid

    id2row = {str(r["id"]).strip(): r for _, r in df_m.iterrows()} #markables by id
    by_sent = df_m.groupby("sent_id") #markables by sentence


    #get the first index 
    def word_index(span):
        if not isinstance(span, str):
            return None
        m = re.search(r"word_(\d+)", span)
        return int(m.group(1)) if m else None

    negatives = []
    for _, row in df_pos.iterrows():
        gold_ant_id = str(row["antecedent_id"]).strip()
        gold_ref_id = str(row["reference_id"]).strip()

        gold_ant = id2row.get(gold_ant_id)
        gold_ref = id2row.get(gold_ref_id)
        if gold_ant is None or gold_ref is None:
            continue

        ant_sent = gold_ant["sent_id"]
        if ant_sent not in by_sent.groups:
            continue

        ref_emb = get_embedding_for_span(" ".join(gold_ref["word_ids"]))
        if ref_emb is None:
            continue

        ref_word_index = word_index(gold_ref["span"])

        # look for other makrables in the same sentences to be used as antecedents 
        for _, cand in by_sent.get_group(ant_sent).iterrows():
            cand_id = str(cand["id"]).strip()
            cand_text = str(cand["words"]).strip()

            if cand_id == gold_ant_id:
                continue
            if cand_text.lower() == str(row["antecedent"]).lower().strip():
                continue

            # skip pronouns as antecedents
            if cand_text.lower() in pronouns:
                continue

            # exclude the second person pronouns
            if not is_valid_antecedent(cand_text):
                continue

            # skip the same cluster
            if full_coref_clusters and markable2cluster.get(cand_id) == markable2cluster.get(gold_ant_id):
                continue  

            # no candidates after the reference
            cand_word_index = word_index(cand["span"])
            if cand_word_index is None or ref_word_index is None:
                continue
            if cand_word_index >= ref_word_index:
                continue

            cand_emb = get_embedding_for_span(" ".join(cand["word_ids"]))
            if cand_emb is None:
                continue

            cos_sim = cosine_similarity(
                cand_emb.unsqueeze(0),
                ref_emb.unsqueeze(0)
            )[0][0].item()

            raw_fn = markables.get(cand_id, {}).get("gram_fnc", "UNK")
            gram_fn = raw_fn if raw_fn in {"subj", "obj"} else "other"

            negatives.append({
                "Sent_1": row["Sent_1"],
                "Sent_2": row["Sent_2"],
                "antecedent": cand["words"],     # wrong antecedent
                "reference": row["reference"],   # gold reference
                "antecedent_id": cand_id,
                "reference_id": gold_ref_id,
                "antecedent_function": gram_fn,
                "cosine_sim": cos_sim,
                "embedding_ante": cand_emb,
                "embedding_ref": ref_emb,
                "label": 0
            })

    return pd.DataFrame(negatives)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizerFast, BertModel
from sklearn.metrics import classification_report
import numpy as np
import pandas as pd
```

```python
# the classifioer was build based on a mix of methods 
#https://docs.pytorch.org/tutorials/beginner/basics/data_tutorial.html 
#https://medium.com/@mn05052002/building-a-simple-mlp-from-scratch-using-pytorch-7d50ca66512b
#https://github.com/GLAZERadr/Multi-Layer-Perceptron-Pytorch/blob/main/model/MultiLayerPerceptron.ipynb



class CorefDataset(Dataset):
    def __init__(self, ids1, mask1, span1, ids2, mask2, span2, labels):
        self.ids1 = ids1 # antecedents
        self.mask1 = mask1
        self.span1 = span1
        self.ids2 = ids2
        self.mask2 = mask2 #references
        self.span2 = span2
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            'ids1': torch.tensor(self.ids1[idx], dtype=torch.long),
            'mask1': torch.tensor(self.mask1[idx], dtype=torch.long),
            'span1': torch.tensor(self.span1[idx], dtype=torch.float),
            'ids2': torch.tensor(self.ids2[idx], dtype=torch.long),
            'mask2': torch.tensor(self.mask2[idx], dtype=torch.long),
            'span2': torch.tensor(self.span2[idx], dtype=torch.float),
            'label': torch.tensor(self.labels[idx], dtype=torch.float)
        }
```

```python
#https://www.kaggle.com/code/hughiephan/classification-with-mean-pooling-bert 

#mean pooling of spans
def masked_mean(embeddings, span_mask):
    span_mask = span_mask.unsqueeze(-1)
    summed = (embeddings * span_mask).sum(dim=1)
    counts = span_mask.sum(dim=1) + 1e-6
    return summed / counts
```

```python
class CorefClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.bert = BertModel.from_pretrained("bert-base-uncased") #finetune the model
        self.mlp = nn.Sequential(
            nn.Linear(4 * 768, 256), # the linearity of the model was changed for the larger versions
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 1)
        )

    def forward(self, ids1, mask1, span1, ids2, mask2, span2):
        emb1 = self.bert(input_ids=ids1, attention_mask=mask1).last_hidden_state
        emb2 = self.bert(input_ids=ids2, attention_mask=mask2).last_hidden_state

        vec1 = masked_mean(emb1, span1)
        vec2 = masked_mean(emb2, span2)

        concat = torch.cat([vec1, vec2, torch.abs(vec1 - vec2), vec1 * vec2], dim=-1)
        return self.mlp(concat).squeeze(1)
```

```python
def train_loop(model, dataloader, optimizer, loss_fn, device):
    model.train()
    total_loss = 0
    for batch in dataloader:
        optimizer.zero_grad()
        for k in batch:
            batch[k] = batch[k].to(device)
        out = model(batch['ids1'], batch['mask1'], batch['span1'],
                    batch['ids2'], batch['mask2'], batch['span2'])
        loss = loss_fn(out, batch['label'])
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)
```

```python
def evaluate(model, dataloader, device):
    model.eval()
    preds, trues = [], []
    with torch.no_grad():
        for batch in dataloader:
            for k in batch:
                batch[k] = batch[k].to(device)
            out = model(batch['ids1'], batch['mask1'], batch['span1'],
                        batch['ids2'], batch['mask2'], batch['span2'])
            preds.extend((out.sigmoid() >= 0.5).long().cpu().numpy())
            trues.extend(batch['label'].long().cpu().numpy())
    return classification_report(trues, preds, digits=3)

```

```python
from sklearn.model_selection import train_test_split
import numpy as np

df = pd.read_csv("all_data_balanced.csv")
df = df.dropna(subset=["Sent_1", "Sent_2", "antecedent", "reference", "label"]).reset_index(drop=True)
y = df["label"].astype(int).values

# use the same function as before fon finding span
def find_span(sent, mention):
    s = sent.lower(); m = mention.lower().strip()
    i = s.find(m)
    if i < 0:
        return None
    return (i, i + len(m))

tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")

#here the help of GPT was needed
def encode_with_span(sent_list, span_text_list, max_len=256):
    enc = tokenizer(
        sent_list,
        return_offsets_mapping=True,
        padding=True,
        truncation=True,
        max_length=max_len,
        return_tensors="np"
    )
    offsets = enc["offset_mapping"]
    span_masks = np.zeros_like(enc["input_ids"])

    # iterrate trhough the span characters in the sentences
    for i, (sent, span_text) in enumerate(zip(sent_list, span_text_list)):
        ch_span = find_span(sent, span_text)
        if ch_span is None:
            span_masks[i, 0] = 1
            continue
        ch_start, ch_end = ch_span

        # iterate over the offsets
        for j, (a, b) in enumerate(offsets[i]):
            if a == b == 0: continue
            if not (b <= ch_start or a >= ch_end):
                span_masks[i, j] = 1
        if span_masks[i].sum() == 0:
            span_masks[i, 0] = 1
    enc.pop("offset_mapping") # drop them 

    return enc["input_ids"], enc["attention_mask"], span_masks

# sentence encoding
ids1, mask1, span1 = encode_with_span(df["Sent_1"].tolist(), df["antecedent"].tolist(), max_len=256)
ids2, mask2, span2 = encode_with_span(df["Sent_2"].tolist(), df["reference"].tolist(), max_len=128)

# datasplit
(y_train, y_test,
 span1_tr, span1_te,
 span2_tr, span2_te,
 ids1_tr, ids1_te,
 mask1_tr, mask1_te,
 ids2_tr, ids2_te,
 mask2_tr, mask2_te) = train_test_split(
    y, span1, span2, ids1, mask1, ids2, mask2,
    test_size=0.2, random_state=42, stratify=y
)
```

```python

# we connect word to sentences, before we have done the character to words and started like that fro the reference chains but it didnt work
def build_word_to_sentence(dfsent):
    """
    map from word_id to sentence id
    using the sentence spans in dfsent
    """
    unique_sent_ids = sorted(dfsent["orderid"].unique())
    sent_id_map = {orig: idx for idx, orig in enumerate(unique_sent_ids)}

    word_to_sentence = {}

    for _, row in dfsent.iterrows():
        sentence_id = sent_id_map[row["orderid"]]
        span = row["span"]
        if not span:
            continue

        parts = span.split(",")
        for part in parts:
            part = part.strip()
            if ".." in part:
                try:
                    start_id, end_id = part.split("..")
                    start = int(start_id.replace("word_", ""))
                    end = int(end_id.replace("word_", ""))
                    for i in range(start, end + 1):
                        word_to_sentence[f"word_{i}"] = sentence_id
                except ValueError:
                    continue
            else:
                word_to_sentence[part] = sentence_id

    return word_to_sentence
```

```python
def build_sentence_text_map(word_to_sentence, words):
    """
    Build mapping from sentence_id -> sentence text.
    Input: word_to_sentence mapping and words dict.
    """
    sentence_words = defaultdict(list)

    #group words ins sentences
    for wid, word in words.items():
        sent_idx = word_to_sentence.get(wid)
        if sent_idx is not None:
            sentence_words[sent_idx].append((int(wid.split("_")[1]), word))

    #sort them numerically 
    sentence_index_to_text = {}
    for idx, word_list in sentence_words.items():
        sorted_words = [w for _, w in sorted(word_list)]
        sentence_index_to_text[idx] = " ".join(sorted_words)

    return sentence_index_to_text
```

```python
def load_markables_and_words(tree, branch):
    markables = {}
    for m in tree:
        if "id" in m.attrib:
            markables[m.attrib["id"]] = m.attrib

    words = {}
    for w in branch.iter():
        if w.tag.endswith("word") and "id" in w.attrib:
            words[w.attrib["id"]] = w.text
    return markables, words
```

```python
#same principles get the clsusters

def get_coref_clusters(markables):
    clusters = {}
    for mid, info in markables.items():
        if info.get("category") == "abstract":
            continue
        coref_set = info.get("coref_set")
        ref_type = info.get("ref_type")
        if coref_set and ref_type == "phrase":
            clusters.setdefault(coref_set, []).append(mid)
    return clusters
```

```python
#create the chains
def build_full_chains(coref_clusters, markables):
    full = {}
    for set_id, mids in coref_clusters.items():
        collected = set(mids)
        for mid in mids:
            collected.update(trace_antecedents(mid, markables))
        # sort by numeric id to get a stable order along the text
        full[set_id] = sorted(collected, key=lambda m: int(m.split("_")[1]))
    return full
```

```python
# map the sentences to number 
def build_sentence_text_map(word_to_sentence, words):
    sentence_words = defaultdict(list)

    #look for the sentence of the word
    for wid, word in words.items():
        sent_idx = word_to_sentence.get(wid)
        if sent_idx is not None:
            sentence_words[sent_idx].append((int(wid.split("_")[1]), word))

    #create the sentences
    sentence_index_to_text = {}
    for idx, word_list in sentence_words.items():
        sorted_words = [w for _, w in sorted(word_list)]
        sentence_index_to_text[idx] = " ".join(sorted_words)

    return sentence_index_to_text
```

```python
def trace_antecedents(mid, markables):
    """
    go backwards to trace the antecedents
    """
    chain = []
    current = mid
    while current and current not in chain:
        chain.append(current)
        info = markables.get(current)
        if not info:
            break
        next_ = info.get("single_phrase_antecedent") or (
            info.get("multiple_phrase_antecedents", "").split(";")[0]
            if info.get("multiple_phrase_antecedents") else None
        )
        current = next_
    return chain


def build_full_coref_clusters(coref_clusters, markables):
    """
build the full coreference chains
    """
    full_clusters = {}
    for set_id, mids in coref_clusters.items():
        collected = set(mids)
        for mid in mids:
            collected.update(trace_antecedents(mid, markables))

        sorted_chain = sorted(
            [m for m in collected if m in markables],
            key=lambda m: int(m.replace("markable_", ""))
        )
        full_clusters[set_id] = sorted_chain
    return full_clusters
```

```python
def build_filtered_pairs(full_coref_clusters, markables, words, word_to_sentence, sentence_index_to_text, pronouns, step=3):
    """
    create positive pairs based on the steps back that we want to go 
    """

    rows = []

    for set_id, chain in full_coref_clusters.items():
        if len(chain) < step + 1:
            continue  

        for i in range(len(chain) - step):  
            antecedent_id = chain[i]
            reference_id  = chain[i + step]

            # skip possible mistaken words from spans
            if antecedent_id not in markables or reference_id not in markables:
                continue

            ante_span = markables[antecedent_id].get("span")
            ref_span  = markables[reference_id].get("span")
            if not ante_span or not ref_span:
                continue

            ante_emb = get_embedding_for_span(ante_span)
            ref_emb  = get_embedding_for_span(ref_span)
            if ante_emb is None or ref_emb is None:
                continue

            ante_text = span_to_text(ante_span, words)
            ref_text  = span_to_text(ref_span, words)

            # pronominal references
            if ref_text.lower() not in pronouns:
                continue

            #kskip pronouns as antecedents
            if ante_text.lower() in pronouns:
                continue

            # sentence indices
            def get_sentence_for_markable(markable_id):
                span = markables[markable_id]["span"]
                first_part = span.split(",")[0].strip()
                first_word = first_part.split("..")[0] if ".." in first_part else first_part
                return word_to_sentence.get(first_word)

            ante_sent = get_sentence_for_markable(antecedent_id)
            ref_sent  = get_sentence_for_markable(reference_id)

            if ante_sent is None or ref_sent is None:
                continue
            if ref_sent < ante_sent:  # antecedent must be before the reference
                continue

            # grammatical functions
            raw_fn = markables[antecedent_id].get("gram_fnc", "UNK")
            ante_function = raw_fn if raw_fn in {"subj", "obj"} else "other"

            ref_function = markables[reference_id].get("gram_fnc", "UNK")
            ref_function = ref_function if ref_function in {"subj", "obj"} else "other"

            # cosine similarity between atecedent and 
            cos_sim = F.cosine_similarity(
                ante_emb.unsqueeze(0), ref_emb.unsqueeze(0)
            ).item()

            rows.append({
                "chain_id": set_id,
                "antecedent_id": antecedent_id,
                "reference_id": reference_id,
                "antecedent_text": ante_text,
                "reference_text": ref_text,
                "sent_1": ante_sent,
                "sent_2": ref_sent,
                "sent_1_text": sentence_index_to_text.get(ante_sent, ""),
                "sent_2_text": sentence_index_to_text.get(ref_sent, ""),
                "distance": ref_sent - ante_sent,
                "embedding_ante": ante_emb,
                "embedding_ref": ref_emb,
                "cosine_sim": cos_sim,
                "antecedent_function": ante_function,
                "reference_function": ref_function,
                "label": 1
            })

    return pd.DataFrame(rows)
```

```python
# we have to exlude the 1st and 2nd person samples

FIRST_SECOND = {
    "i", "me", "my", "mine", "we", "us", "our", "ours",
    "you", "your", "yours", "yourself", "yourselves"
}

# again we chaek about valid anyecedents
def is_valid_antecedent(text: str) -> bool:
    """
    Return True if antecedent is acceptable:
    - Reject pure 1st/2nd person pronouns (you, I, we, etc.)
    - Allow spans like 'your ovaries' because they contain a content word.
    """
    tokens = text.lower().split()
    if len(tokens) == 1 and tokens[0] in FIRST_SECOND:
        return False
    if all(tok in FIRST_SECOND for tok in tokens):
        return False
    return True
```

```python
def build_negatives_wrong_antecedent_close(
    df_pos, markables, words,
    word_to_sentence, sentence_index_to_text,
    pronouns, full_coref_clusters,
    window=1
):
    """
    Same reference wrong antecedent. same principle like the previous negative but tracing back to the real antecedent
    and within the window of 1 we try to find a near noun phrase  
    """


    markable2cluster = {}
    for cid, mids in full_coref_clusters.items():
        for m in mids:
            markable2cluster[m] = cid

    negatives = []

    for _, row in df_pos.iterrows():
        gold_ant_id = row["antecedent_id"]
        gold_ref_id = row["reference_id"]

        gold_ant_sent = row["sent_1"]
        ref_sent = row["sent_2"]

        # not pronoun to pronoun
        if row["reference_text"].lower() not in pronouns:
            continue

        gold_cluster = markable2cluster.get(gold_ant_id, None)

        for mid, info in markables.items():
            if mid == gold_ant_id:
                continue  # skip gold antecedent
            if markable2cluster.get(mid) == gold_cluster:
                continue  # skip same cluster (true coref)

            span = info.get("span")
            if not span:
                continue

            # sentence for candidate antecedent
            first_part = span.split(",")[0].strip()
            first_word = first_part.split("..")[0] if ".." in first_part else first_part
            cand_sent = word_to_sentence.get(first_word)
            if cand_sent is None:
                continue

            # keep the options close to the real np
            if abs(cand_sent - gold_ant_sent) > window:
                continue

            cand_text = span_to_text(span, words)
           #skip first seconf 
            if cand_text.lower() in pronouns:
                continue

            if not is_valid_antecedent(cand_text):
                continue

            cand_emb = get_embedding_for_span(span)
            if cand_emb is None:
                continue

            # grammatical functions
            raw_fn_ant = markables[mid].get("gram_fnc", "UNK")
            ant_function = raw_fn_ant if raw_fn_ant in {"subj", "obj"} else "other"

            raw_fn_ref = markables[gold_ref_id].get("gram_fnc", "UNK")
            ref_function = raw_fn_ref if raw_fn_ref in {"subj", "obj"} else "other"

            # cosine similarity
            cos_sim = F.cosine_similarity(
                cand_emb.unsqueeze(0),
                row["embedding_ref"].unsqueeze(0)
            ).item()

            negatives.append({
                "chain_id": row["chain_id"],
                "antecedent_id": mid,
                "reference_id": gold_ref_id,
                "antecedent_text": cand_text,
                "reference_text": row["reference_text"],
                "sent_1": cand_sent,
                "sent_2": ref_sent,
                "sent_1_text": sentence_index_to_text.get(cand_sent, ""),
                "sent_2_text": row["sent_2_text"],
                "distance": abs(ref_sent - cand_sent),
                "embedding_ante": cand_emb,
                "embedding_ref": row["embedding_ref"],
                "cosine_sim": cos_sim,
                "antecedent_function": ant_function,
                "reference_function": ref_function,
                "label": 0
            })

    return pd.DataFrame(negatives)
```

```python
#tak the average of the columns
def avg_cosine_from_csv(pos_file, neg_file):
    df_pos = pd.read_csv(pos_file)
    df_neg = pd.read_csv(neg_file)
    pos_avg = df_pos["cosine_sim"].mean()
    neg_avg = df_neg["cosine_sim"].mean()
    return pos_avg, neg_avg

#Short
pos_short, neg_short = avg_cosine_from_csv("all_data_balanced_positive.csv", "all_data_balanced_negative.csv")
# Medium
pos_medium, neg_medium = avg_cosine_from_csv("data_mark3_positives.csv", "data_mark3_negatives.csv")
# Long
pos_long, neg_long = avg_cosine_from_csv("data_mark5_positives.csv", "data_mark5_negatives.csv")

print("Short:", pos_short, neg_short)
print("Medium:", pos_medium, neg_medium)
print("Long:", pos_long, neg_long)
```

```python
#define the labes
labels = ["Short", "Medium", "Long"]
pos_vals = [pos_short, pos_medium, pos_long]
neg_vals = [neg_short, neg_medium, neg_long]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(8,5))

# Pplot with bars red and green
ax.bar(x - width/2, pos_vals, width, label="Positive (coref)", color="red")
ax.bar(x + width/2, neg_vals, width, label="Negative (non-coref)", color="green")

ax.set_ylabel("Average Cosine Similarity")
ax.set_title("Average Cosine Similarity of Coreferent vs Non-coreferent Pairs BERT-BASE")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
